### Notebook per ottenere la visualizzazione relativa al gap dei voti 

```python
import json
import pandas as pd
```

```python
df = pd.read_json("imdb_dataset_parsed.json")
```

```python
df.head(3)
```

```python
# Normalizzazione dei punteggi di voti
df["metascore"] = df["metascore"] / 10
# Creazione dataset per ottnere il file relativo al gap
df_score = pd.DataFrame(df.dropna())
# Calcolo discordanza
df_score["score_diff"] = (df_score["metascore"] - df["users_rating"] ) / df_score["metascore"]
# Correzione campo 'voti'
df_score["votes"] = df_score["votes"].str.replace(",", "").astype(int)
```

```python
# Selezione di film che hanno più di 150'000 voti, dal 2009 in poi e che hanno una discordanza maggiore del 10%
df_score = df_score[df_score["votes"] > 150000]
df_score = df_score[(df_score["score_diff"]) > 0.10]
df_score =  df_score[df_score["year"] > 2009]
```
